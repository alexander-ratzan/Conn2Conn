```python
# Small defaults for quick local iteration
config_override = {
    "model": {
        "layer_num": 2,
        "hidden_dim": 128,
        "decoder_dim": 64,
        "dropout": 0.1,
        "edge_dropout": 0.1,
        "reg": 1e-5,
        "use_volume": True,
        "use_spatial": True,
        "use_r2t": True,
    },
    "trainer": {
        "max_epochs": 750,
        "batch_size": 16,
        "lr": 2e-3,
        "loss_type": "mse",
        "log_every": 2,
    },
}

results = []

for seed in range(10):
    sim = Sim(
        model_name="NodalGNN",
        config_overrides=config_override,
        source="SC",
        target="FC",
        parcellation="Glasser",
        shuffle_seed=seed,
        data_load_mode="precomputed",
    )
    run_out = sim._run_learned_single(
        mode="prod",
        save_checkpoint=True,
        config_override=config_override,
    )
    # Store or print the base metrics for each run
    base_metrics = run_out["test_metrics"]["base_metrics"]
    results.append({"seed": seed, "base_metrics": base_metrics})

# Optionally, print all results
for res in results:
    print(f"Seed: {res['seed']}  Base Metrics: {res['base_metrics']}")
```
