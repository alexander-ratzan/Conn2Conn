# Track Krakencoder Precomputed

Notebook for logging `Krakencoder_precomputed` runs to W&B across multiple seeds.

```python
import importlib
import main
import models.config
import models.models
from main import Sim

importlib.reload(models.models)
importlib.reload(models.config)
importlib.reload(main)
```

```python
# Basic run settings
source = "SC"
target = "FC"
parcellation = "Glasser"
seeds = list(range(10))
kraken_predictions_dir = None  # None -> use repo default krakencoder/example_data

# Set to a smaller list like [0] or [0, 1] for a quick smoke test first.
```

```python
# Optional single-seed smoke test before launching the full loop
seed = seeds[0]

sim = Sim(
    model_name="Krakencoder_precomputed",
    source=source,
    target=target,
    parcellation=parcellation,
    shuffle_seed=seed,
)

run_out = sim._run_closed_form_single(
    mode="prod",
    save_checkpoint=False,
    config_override={
        "model": {
            "kraken_predictions_dir": kraken_predictions_dir,
        }
    },
    wandb_name=f"Krakencoder_precomputed_seed{seed}",
    wandb_group=f"Krakencoder_precomputed_{source}_{parcellation}",
    wandb_tags=["krakencoder", "precomputed", source, parcellation],
    wandb_metadata={
        "seed": seed,
        "source": source,
        "target": target,
        "parcellation": parcellation,
    },
)

run_out["test_metrics"]["base_metrics"]
```

```python
# Full 10-seed tracking loop
all_results = {}

for seed in seeds:
    print(f"Running seed {seed}...")
    sim = Sim(
        model_name="Krakencoder_precomputed",
        source=source,
        target=target,
        parcellation=parcellation,
        shuffle_seed=seed,
    )

    run_out = sim._run_closed_form_single(
        mode="prod",
        save_checkpoint=False,
        config_override={
            "model": {
                "kraken_predictions_dir": kraken_predictions_dir,
            }
        },
        wandb_name=f"Krakencoder_precomputed_seed{seed}",
        wandb_group=f"Krakencoder_precomputed_{source}_{parcellation}",
        wandb_tags=["krakencoder", "precomputed", source, parcellation],
        wandb_metadata={
            "seed": seed,
            "source": source,
            "target": target,
            "parcellation": parcellation,
        },
    )

    all_results[seed] = run_out["test_metrics"]["base_metrics"]

all_results
```
